```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
dataku.head()
```

```python
dataku.corr() #tabel berformat matrix
```

```python
sns.heatmap(dataku.corr());
```

```python
sns.heatmap(dataku.corr(), annot=True);
```

```python
sns.heatmap(dataku.corr(), annot=True, cmap='BrBG');
```

```python
dataku2 = pd.read_csv('kapal_titanic.csv', index_col='deck')
dataku2.head()
```

```python
terbang = sns.load_dataset('flights')
```

```python
terbang
```

```python
terbang_pivot = terbang.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(terbang_pivot, cmap='BrBG', linecolor='black', linewidths=1);
```

```python
sns.clustermap(terbang_pivot); # diagram pohon di bawah ini disebut dengan dendogram 
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm'); # diagram pohon di bawah ini disebut dengan dendogram 
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm', standard_scale=1); # diagram pohon di bawah ini disebut dengan dendogram 
```

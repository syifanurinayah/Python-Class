```python
import pandas as pd
```

```python
satu = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                        'one':[1,2,3,4,5],
                        'two':[6,7,8,9,10]})
```

```python
dua = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                        'three':[1,2,3,4,5],
                        'four':[6,7,8,9,10]})
```

```python
baru = pd.merge(satu,dua, how='inner', on='key')
baru
```

```python
satu = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                        'A':['A0','A1','A2','A3'],
                        'B':['B0','B1','B2','B3']})

dua = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                        'C':['C0','C1','C2','C3'],
                        'D':['D0','D1','D2','D3']})
```

```python
satu
```

```python
dua
```

```python
baru2 = pd.merge(satu,dua, how='left', on=['key1','key2'])
```

```python
baru2
```

```python
baru2 = pd.merge(satu,dua, how='right', on=['key1','key2'])
```

```python
baru2
```

```python
baru2 = pd.merge(satu,dua, how='outer', on=['key1','key2'])
```

```python
baru2
```

```python
tiga = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                        'A':['A0','A1','A2','A3'],
                        'B':['B0','B1','B2','B3']})

empat = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                     'key2':['K0','K0','K0','K0'],
                        'C':['C0','C1','C2','C3'],
                        'D':['D0','D1','D2','D3']})
```

```python
tiga
```

```python
empat
```

```python
baru3 = pd.merge(tiga,empat, how='left', on=['key1','key2'])
baru3
```

```python
baru3 = pd.merge(tiga,empat, how='right', on=['key1','key2'])
baru3
```

```python
baru3 = pd.merge(tiga,empat, how='outer', on=['key1','key2'])
baru3
```

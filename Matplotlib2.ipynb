```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(3,20,3)
y = x**2
```

```python
x
```

```python
y
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y, color='purple')
axes1.set_xlabel('Sumbu X')
axes1.set_ylabel('Sumbu Y')
axes1.set_title('Judul Grafik');
axes2.plot(y,x)
axes2.set_xlabel('Sumbu X')
axes2.set_ylabel('Sumbu Y')
axes2.set_title('Judul Grafik');

plt.show()
```

```python
# versi 2
fig, axes = plt.subplots(1,2)

#plot for the first axes 
axes[0].plot(x,y)
axes[0].set_xlabel('sumbu x1')
axes[0].set_ylabel('sumbu y1')
axes[0].set_title('Judul 1')

#plot for the second axes
axes[1].plot(x,y)
axes[1].set_xlabel('sumbu x2')
axes[1].set_ylabel('sumbu y2')
axes[1].set_title('Judul 2')

fig.tight_layout()
```

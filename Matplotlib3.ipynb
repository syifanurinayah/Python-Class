```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,20,0.5)
y = x**2
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi=100)
axes[0].plot(x,y)
axes[0].set_title('Grafik 1')

axes[1].plot(x,y)
axes[1].set_title('Grafik 2')

plt.tight_layout()

plt.show()
```

```python
fig.savefig('gambarku2.png', dpi=300)
```

# menambahkan legenda

```python
fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Panngkat 2')
ax.plot(x,x**3, label='Pangkat 3')
ax.plot(x,x**4, label='Pangkat 4')

ax.legend()
plt.show()
```

```python
fig = plt.figure(dpi=200)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='Panngkat 2', lw=3, ls='-', marker='o', markersize=5, mfc='purple', mew=1, mec='blue')
ax.plot(x,x**3, label='Pangkat 3', lw=2, ls='--', marker='*', markersize=5, mfc='purple', mew=1, mec='blue')
ax.plot(x,x**4, label='Pangkat 4',lw=3, ls='-.', marker='+', markersize=5, mfc='purple', mew=1, mec='blue')

ax.legend(loc=0)
ax.set_xlim(0,12.5)
ax.set_ylim(0,2500)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
fig.subplots_adjust(top=0.8)
ax1 = fig.add_subplot(211)
ax1.set_ylabel('volts')
ax1.set_title('a sine wave')

t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = ax1.plot(t, s, color='blue', lw=2)

# Fixing random state for reproducibility
np.random.seed(19680801)

ax2 = fig.add_axes([0.15, 0.1, 0.7, 0.3])
n, bins, patches = ax2.hist(np.random.randn(1000), 50,
                            facecolor='yellow', edgecolor='yellow')
ax2.set_xlabel('time (s)')

plt.show()
```

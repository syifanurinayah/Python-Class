```python
import pandas as pd
```

```python
satu = pd.DataFrame({'A':['a1','a2','a3','a4','a5'],
                        'B':['b1','b2','b3','b4','b5'],
                        'C':['c1','c2','c3','c4','c5'],
                        'D':['d1','d2','d3','d4','d5']},
                       index=[0,1,2,3,4])
```

```python
dua = pd.DataFrame({'A':['a6','a7','a8','a9','a10'],
                        'B':['b6','b7','b8','b9','b10'],
                        'C':['c6','c7','c8','c9','c10'],
                        'D':['d6','d7','d8','d9','d10']},
                       index=[5,6,7,8,9])
```

```python
tiga = pd.DataFrame({'A':['a11','a12','a13','a14','15'],
                        'B':['b11','b12','b13','b14','b15'],
                        'C':['c11','c12','c13','c14','c15'],
                        'D':['d11','d12','d13','d14','d15']},
                       index=[10,11,12,13,14])
```

```python
bisnis = pd.DataFrame({'perusahaan':['Telkom','Telkom','Pertamnina','Indosat','Pertamina','BCA','KFC','KFC'],
                          'karyawan':['joko','budi','wati','soni','teti','juki','jenab','yuyu'],
                          'usia':[20,25,30,30,27,28,35,39]})
```

```python
bisnis
```

```python
bisnis.groupby('perusahaan')
```

```python
bisnis.groupby('perusahaan').sum()
```

```python
bisnis.groupby('perusahaan').count()
```

```python
bisnis.groupby('perusahaan').mean()
```

```python
bisnis.groupby('perusahaan').usia.min()
```

```python
bisnis.groupby('perusahaan').describe()
```

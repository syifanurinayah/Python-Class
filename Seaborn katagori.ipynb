```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
```

```python
dataku.head()
```

```python
dataku.describe(include='O')
```

```python
# barplot (histogram)

sns.barplot(x='sex', y='fare', data=dataku, estimator=sum);
```

```python

sns.barplot(x='sex', y='fare', data=dataku, estimator=np.std);
```

```python
# countplot
sns.countplot(x='sex',data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', order=['A','B','C','D','E','F','G'], data=dataku);
```

```python
#boxplot
sns.boxplot(data=dataku, x='sex', y='fare');
```

```python
#boxplot
sns.boxplot(data=dataku, x='sex', y='fare',hue_order=['A','B','C','D','E','F','G'], hue='deck');
```

```python
# violin plot
sns.violinplot(data=dataku, x='sex', y='fare');
```

```python
sns.violinplot(data=dataku, x='deck', y='fare',  hue='sex', order=['A','B','C','D','E','F','G'], split=True);
```

```python
# stripplot
sns.stripplot(data=dataku, x='deck', y='fare',  hue='sex', order=['A','B','C','D','E','F','G'], jitter=True);
```

```python
# swarmplot

sns.swarmplot(data=dataku, x='deck', y='fare',  hue='sex', order=['A','B','C','D','E','F','G']);
```

```python
sns.violinplot(data=dataku, x='deck', y='fare',  hue='sex', order=['A','B','C','D','E','F','G']);
sns.swarmplot(data=dataku, x='deck', y='fare',  hue='sex', order=['A','B','C','D','E','F','G'], color='black');
```
